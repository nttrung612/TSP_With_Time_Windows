## **Backtrack with Sorting Time**

```python
import time
import sys

neighbor = {}
for i in range(n + 1):
    neighbor[i] = sorted(
        [(j, t[i][j]) for j in range(n + 1) if i != j],
        key=lambda x : x[1]
    )

node_visited = 0

def check(j, cur_time, vis, prev_node):
    if vis[j]:
        return False
    if cur_time + t[prev_node][j] > l[j]:
        return False
    return True

def backtrack(i, f, vis, cur_time, path, best_ans, best_path, start_time):
    global node_visited
    node_visited += 1
    if time.time() - start_time > 30:
        # print(n)
        # print(" ".join(map(str, best_path)))
        print("TLE")
        sys.exit(0)
    if i == n:
        if best_ans[0] > f:
            best_ans[0] = f
            best_path[:] = path[:]
            # print(best_ans[0])
        return
    prev_node = path[-1] if path else 0
    for j, travel_time in neighbor[prev_node]:
        if check(j, cur_time, vis, prev_node):
            vis[j] = True
            f += travel_time
            cur_time += travel_time
            new_cur_time = max(cur_time, e[j]) + d[j]
            path.append(j)
            if f + (n - i) * tmin < best_ans[0]:
                backtrack(i + 1, f, vis, new_cur_time, path, best_ans, best_path, start_time)
            vis[j] = False
            f -= travel_time
            cur_time -= travel_time
            path.pop()

best_ans = [float('inf')]
best_path = []
vis = [False] * (n + 1)
start_time = time.time()
backtrack(0, 0, vis, 0, [], best_ans, best_path, start_time)
print(best_ans[0])
print(n)
print(" ".join(map(str, best_path)))
print(f"Nodes visited: {node_visited}")
```

## **Dynamic Programming**

```python
with open("input.txt") as f:
    lines = f.readlines()
n = int(lines[0])
e, l, d = [0], [0], [0]
for i in range(1, n + 1):
    e1, l1, d1 = map(int, lines[i].split())
    e.append(e1)
    l.append(l1)
    d.append(d1)
t = [list(map(int, lines[i].split())) for i in range(n + 1, 2*n + 2)]
```

```python
dp = [[float('inf')] * (1 << (n + 1)) for _ in range(n + 1)]
dp[0][1] = 0

prev = [[(-1, -1)] * (1 << (n + 1)) for _ in range(n + 1)]

for mask in range(1 << (n + 1)):
    for i in range(n + 1):
        if not (mask & (1 << i)) or dp[i][mask] == float('inf'):
            continue
        for j in range(1, n + 1):  # only customers
            if mask & (1 << j):
                continue
            arrive = dp[i][mask] + t[i][j]
            start_service = max(arrive, e[j])
            if start_service <= l[j]:
                new_mask = mask | (1 << j)
                dp[j][new_mask] = min(dp[j][new_mask], start_service + d[i])
                prev[j][new_mask] = (i, mask)
```

```python
def trace_path(dp, prev, n):
    full_mask = (1 << (n + 1)) - 1
    end_node = -1
    min_time = float('inf')

    for i in range(1, n + 1):
        if dp[i][full_mask] < min_time:
            min_time = dp[i][full_mask]
            end_node = i

    if end_node == -1:
        return []

    path = []
    curr = end_node
    mask = full_mask

    while curr != -1:
        path.append(curr)
        curr, mask = prev[curr][mask]

    path.reverse()
    return path[1:]

path = trace_path(dp, prev, n)
print(n)
print(" ".join(map(str, path)))
```

## **Integer Programming**

```python
with open("input.txt") as f:
    lines = f.readlines()
n = int(lines[0])
e, l, d = [0], [0], [0]
for i in range(1, n + 1):
    e1, l1, d1 = map(int, lines[i].split())
    e.append(e1)
    l.append(l1)
    d.append(d1)
t = [list(map(int, lines[i].split())) for i in range(n + 1, 2*n + 2)]
```

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SAT')

time_limit_ms = 30 * 1000
solver.set_time_limit(time_limit_ms)

x = {(i, j) : solver.IntVar(0, 1, f'x[{i},{j}]') for i in range(n+1) for j in range(n+1)}
time = [solver.IntVar(e[i], l[i], f't[{i}]') for i in range(n+1)]

solver.Minimize(sum(x[i, j] * t[i][j] for i in range(n+1) for j in range(n+1) if i != j))

# every node 1 - n is visited exactly once
for i in range(1, n + 1):
    solver.Add(sum(x[j, i] for j in range(n + 1) if j != i) == 1)

# node 0 must visit exactly one node 1 - n
solver.Add(sum(x[0, i] for i in range(1, n + 1)) == 1)

# outcoming from each node 1 - n <= 1
for i in range(1, n + 1):
    solver.Add(sum(x[i, j] for j in range(1, n + 1) if j != i) <= 1)

# time constraints
t0 = 0
M = 1e6
for j in range(1, n + 1):
    solver.Add(M * (1 - x[0, j]) + time[j] >= t0 + t[0][j])
    for i in range(1, n + 1):
        if i != j:
            solver.Add(M * (1 - x[i, j]) + time[j] >= time[i] + d[i] + t[i][j])

status = solver.Solve()
if status == pywraplp.Solver.OPTIMAL:
    print(int(solver.Objective().Value()))
    print(n)
    child = [-1] * (n + 1)
    for i in range(n + 1):
        for j in range(n + 1):
            if i != j and x[i, j].solution_value() == 1:
                child[i] = j
    path = []
    cur = 0
    while cur != -1:
        path.append(cur)
        cur = child[cur]
    print(" ".join(map(str, path[1:])))
    print("Time taken:", solver.wall_time(), "ms")
else:
    print("SAI")
```

## **Constraint Programming**

```python
with open("input.txt") as f:
    lines = f.readlines()
n = int(lines[0])
e, l, d = [0], [0], [0]
for i in range(1, n + 1):
    e1, l1, d1 = map(int, lines[i].split())
    e.append(e1)
    l.append(l1)
    d.append(d1)
t = [list(map(int, lines[i].split())) for i in range(n + 1, 2*n + 2)]
```

```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()

x = {(i, j): model.NewBoolVar(f'x[{i},{j}]') for i in range(n + 1) for j in range(n + 1)}
time = [model.NewIntVar(e[i], l[i], f't[{i}]') for i in range(n + 1)]

model.Minimize(sum(x[i, j] * t[i][j] for i in range(n + 1) for j in range(n + 1) if i != j))

# every node 1 - n is visited exactly once
for i in range(1, n + 1):
    model.Add(sum(x[j, i] for j in range(n + 1) if j != i) == 1)

# node 0 must visit exactly one node 1 - n
model.Add(sum(x[0, i] for i in range(1, n + 1)) == 1)

# outcoming from each node 1 - n <= 1
for i in range(1, n + 1):
    model.Add(sum(x[i, j] for j in range(1, n + 1) if j != i) <= 1)

# time constraints
t0 = 0
for j in range(1, n + 1):
    model.Add(time[j] >= t0 + t[0][j]).OnlyEnforceIf(x[0, j])
    for i in range(1, n + 1):
        if i != j:
            model.Add(time[j] >= time[i] + d[i] + t[i][j]).OnlyEnforceIf(x[i, j])

solver = cp_model.CpSolver()
# solver.parameters.stop_after_first_solution = True
status = solver.Solve(model)
if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    print(int(solver.ObjectiveValue()))
    print(n)
    child = [-1] * (n + 1)
    for i in range(n + 1):
        for j in range(n + 1):
            if i != j and solver.BooleanValue(x[i, j]):
                child[i] = j
    path = []
    cur = 0
    while cur != -1:
        path.append(cur)
        cur = child[cur]
    print(" ".join(map(str, path[1:])))
    print(f"Nodes visited: {solver.NumBranches()}")
    print(f"Time taken in ms: {solver.WallTime() * 1000}")
else:
    print("SAI")
```

```python
def greedy_tsptw(n, e, l, d, t):
    visited = [False] * (n + 1)
    visited[0] = True
    path = [0]
    cur = 0
    cur_time = 0

    for _ in range(n):
        best = -1
        best_arrival = float('inf')
        for j in range(1, n + 1):
            if visited[j]:
                continue
            arrival = cur_time + t[cur][j]
            if arrival <= l[j] and arrival >= e[j] and arrival < best_arrival:
                best = j
                best_arrival = arrival
        if best == -1:
            break  # Cannot extend further
        path.append(best)
        visited[best] = True
        cur_time = max(e[best], best_arrival) + d[best]
        cur = best
    return path
from ortools.sat.python import cp_model

def refine_with_cp_hint(n, e, l, d, t, greedy_path):
    model = cp_model.CpModel()
    solver = cp_model.CpSolver()

    x = {(i, j): model.NewBoolVar(f'x[{i},{j}]') for i in range(n + 1) for j in range(n + 1) if i != j}
    time_var = [model.NewIntVar(e[i], l[i], f'time[{i}]') for i in range(n + 1)]

    # Add constraints
    for i in range(1, n + 1):
        model.Add(sum(x[j, i] for j in range(n + 1) if j != i) == 1)
        model.Add(sum(x[i, j] for j in range(n + 1) if j != i) <= 1)

    model.Add(sum(x[0, j] for j in range(1, n + 1)) == 1)

    for i in range(1, n + 1):
        for j in range(1, n + 1):
            if i != j:
                model.Add(time_var[j] >= time_var[i] + d[i] + t[i][j]).OnlyEnforceIf(x[i, j])
    for j in range(1, n + 1):
        model.Add(time_var[j] >= t[0][j]).OnlyEnforceIf(x[0, j])

    # Objective
    model.Minimize(sum(x[i, j] * t[i][j] for (i, j) in x))

    # --- Add Hint from greedy path ---
    for idx in range(len(greedy_path) - 1):
        i, j = greedy_path[idx], greedy_path[idx + 1]
        if (i, j) in x:
            model.AddHint(x[i, j], 1)

    # Solve with small timeout
    # solver.parameters.max_time_in_seconds = 25
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
        print("✅ CP refinement succeeded.")
        result_path = []
        child = [-1] * (n + 1)
        for (i, j) in x:
            if solver.BooleanValue(x[i, j]):
                child[i] = j
        cur = 0
        while cur != -1:
            result_path.append(cur)
            cur = child[cur]
        print("Refined Path:", result_path)
        return result_path
    else:
        print("❌ CP refinement failed.")
        return None

greedy_path = greedy_tsptw(n, e, l, d, t)
print("Greedy partial path:", greedy_path)
final_path = refine_with_cp_hint(n, e, l, d, t, greedy_path)
```

```python
n, e, l, d, t = read_input()

from ortools.sat.python import cp_model

class FirstSolutionCallback(cp_model.CpSolverSolutionCallback):
    def __init__(self, x, n):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.x = x
        self.n = n
        self.objective = None
        self.path = []

    def on_solution_callback(self):
        print("Feasible solution found")
        self.objective = self.ObjectiveValue()
        child = [-1] * (self.n + 1)
        for i in range(self.n + 1):
            for j in range(self.n + 1):
                if i != j and self.Value(self.x[i, j]):
                    child[i] = j
        cur = 0
        while cur != -1:
            self.path.append(cur)
            cur = child[cur]
        self.StopSearch()

model = cp_model.CpModel()

x = {(i, j): model.NewBoolVar(f'x[{i},{j}]') for i in range(n + 1) for j in range(n + 1)}
time = [model.NewIntVar(e[i], l[i], f't[{i}]') for i in range(n + 1)]

# Objective
model.Minimize(sum(x[i, j] * t[i][j] for i in range(n + 1) for j in range(n + 1) if i != j))

# Constraints
for i in range(1, n + 1):
    model.Add(sum(x[j, i] for j in range(n + 1) if j != i) == 1)
model.Add(sum(x[0, i] for i in range(1, n + 1)) == 1)
for i in range(1, n + 1):
    model.Add(sum(x[i, j] for j in range(1, n + 1) if j != i) <= 1)

t0 = 0
for j in range(1, n + 1):
    model.Add(time[j] >= t0 + t[0][j]).OnlyEnforceIf(x[0, j])
    for i in range(1, n + 1):
        if i != j:
            model.Add(time[j] >= time[i] + d[i] + t[i][j]).OnlyEnforceIf(x[i, j])

solver = cp_model.CpSolver()
callback = FirstSolutionCallback(x, n)
solver.SolveWithSolutionCallback(model, callback)

if callback.path:
    print(int(callback.objective))
    print(n)
    print(" ".join(map(str, callback.path[1:])))
    print(f"Nodes visited: {solver.NumBranches()}")
    print(f"Time taken in ms: {solver.WallTime() * 1000}")
    init_solution = callback.path
```
